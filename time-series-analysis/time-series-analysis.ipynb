### Resampling

Changing the frequency of your timeseries observation

two types of resampling
- **Upsampling** : where you increase the frequency of the samples, such as from minutes to seconds
- **Downsampling** : Where you increase the frequency of the samples, such as from days to months

```python
# downsampling
google['Open'].plot()
google['Open'].resample('QE').mean().plot()
google['Open'].resample('ME').mean().plot()
google['Open'].resample('YE').mean().plot()
```

```python
# Upsampling

google['Close'].resample('12h').interpolate(method='linear').plot()
```

### Rolling window(Smoothing)

TimeSeries data in original format can be quite volatile, especially on smaller aggregation levels. The concept of rolling, or moving average is a useful technique for smoothing time series data.

```python
# rolling window

google['Close'].rolling(3).mean().plot()
```

```python
google['Open'].plot(title='Original')
google['Open'].rolling(5).mean().plot(title='Rolling')
google['Open'].ewm(30).mean().plot(title='EWM')
```
